<h2>House Price Prediction</h2>
<h4>House price prediction can help the developer determine the selling price of a house and can help the customer to arrange the right time to purchase a house. There are three factors that influence the price of a house which include physical conditions, concept and location.</h4>

<ul>
<li>Price - Total cost of house on the basis of requirments</li>
<li>Area - Total area of house in sq.ft.</li>
<li>Bedroom - Total number of bedrooms in house</li>
<li>Bathrooms - Total number of bathrooms in house </li>
<li>Floors - Total Floors in house</li>
<li>Road - House Having mainroad or not</li>
<li>Guestroom - House have guestroom or not</li>
<li>Basement - House is having a basement or not</li>
<li>Geyser - Have geyser Facility or not</li>
<li>Ac - Having Ac Facility or not</li>
<li>PrefAr - House have preference area or not</li>
<li>FurnishingStat - Furnishing status of house</li>
</ul>

<h3>Importing Module</h3>

```python
#importing dependencies

import pandas as pd
import numpy as np
import seaborn as sns
from sklearn import linear_model, metrics
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
from scipy import stats
```

<h3>Importing Dataset</h3>

```python
#Importing housing Dataset
data = pd.read_csv('Housing.csv')
data
```

```python
data.columns=['Price', 'Area','Bedroom','Bathroom', 'Floors', 'Road', 'GuestRoom', 'Basement', 'Geyser', 'Ac', 'ParkingLot', 'PrefAr', 'FurnishingStat']
```

<h3>Perfoming conversion on column which are having categorical type of data</h3>

```python
#Performing Labelencoding conversion
from sklearn.preprocessing import LabelEncoder
categ = ['Road', 'GuestRoom','Basement', 'Geyser', 'Ac', 'PrefAr', 'FurnishingStat']

# Encode Categorical Columns
le = LabelEncoder()
data[categ] = data[categ].apply(le.fit_transform)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats

fig, axs = plt.subplots(ncols=7, nrows=2, figsize=(20, 10))
index = 0
axs = axs.flatten()
for k,v in data.items():
    sns.histplot(y=k, data=data, ax=axs[index])
    index += 1
plt.tight_layout(pad=0.4, w_pad=0.5, h_pad=5.0)
```

<h3>Extracting target column from dataset</h3>

```python
#dividing dataset into target(dependent variable) and dependent varible
target_coln="Price"

#seprate it from dataset
y=data[target_coln]

#Droping from main dataset / input features
x = data.drop(target_coln,axis=1)

y.shape ,x.shape
```

```python
y
```

<h3>Scater plot of each feature on target feature</h3>

```python
column=['Area','Bedroom','Bathroom', 'Floors', 'Road', 'GuestRoom', 'Basement', 'Geyser', 'Ac', 'ParkingLot', 'PrefAr', 'FurnishingStat']
fig, axs = plt.subplots(ncols=7, nrows=2, figsize=(20, 10))

axs = axs.flatten()
for i, k in enumerate(column):
    sns.regplot(y=y, x=x[k], ax=axs[i] ,scatter_kws={"color": "violet"}, line_kws={"color": "blue"})
plt.tight_layout(pad=0.4, w_pad=0.5, h_pad=5.0)
```

<h3>Performing Train Test Split</h3>

```python
x_train , x_test , y_train , y_test = train_test_split(x,y,test_size=0.2 , random_state=7)
```

<h3>Importing Evaluation Module</h3>

```python
#importing Evaluation modules
from sklearn.metrics import r2_score,mean_absolute_error
```

<h3>Importing LinearRegression Algortihm</h3>

```python
# IMPORTING THE MODULE

from sklearn.linear_model import LinearRegression
lr_model = LinearRegression()
```

```python
lr_model.fit(x_train,y_train)
```

```python
lr_predict = lr_model.predict(x_test)
```

```python
lr_predict
```

<h3>Score of LinearRegression model & Mean absolute error</h3>

```python
lr_score = r2_score(y_test,lr_predict)
lr_mae = mean_absolute_error(y_test,lr_predict)

print(lr_score)
lr_mae
```

<h3>Importing RandomForestClassifier Algortihm</h3>

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier(random_state=0)
```

```python
rfc.fit(x_train,y_train)
rfc_predict= rfc.predict(x_test)
rfc_predict
```

<h3>Score of RandomForestClassifier model & Mean absolute error</h3>

```python
score2 = r2_score(y_test,rfc_predict)
mean_a_error = mean_absolute_error(y_test,rfc_predict)

print(score2)
mean_a_error
```

<h3>Hence on the analyzed house price prediction data Linear regression give 67% accuracy while random Forest Classificatio is giving 56% of Accuracy</h3>
